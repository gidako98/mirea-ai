# 01 EDA: synthetic support-ticket data

Goal: inspect data schema, target balance and basic feature distributions before model training.

```
import pandas as pd

data = pd.read_csv('../data/sample_tickets.csv')
data.head()
```

```
data.info()
```

```
data['priority'].value_counts(normalize=True).round(3)
```

```
data.groupby('priority')[['sentiment_score', 'account_age_days']].agg(['mean', 'median']).round(2)
```

## EDA notes

- The target has three classes: `low`, `medium`, `high`.
- The distribution is moderately imbalanced, therefore macro F1 is used alongside accuracy.
- Text, categorical and numeric features all carry useful signal.
- The dataset is synthetic and contains small label noise to mimic imperfect human triage labels.
